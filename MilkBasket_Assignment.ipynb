### Day-on-day Penetration

```python
data.head()
```

```python
data['order_date'].unique()
```

```python
data['category_name'].value_counts()
```

```python
avg_day_to_day_penetration = []
for i in data['category_name'].value_counts().index:
    total_orders_day = data.groupby(['order_date'], as_index=False)['order_id'].count()
    total_orders_day.rename({'order_id':'Total_orders'},axis=1, inplace=True)


    temp = data[data['category_name'] == i][['order_date','category_name','order_id']]


    category_orders_day = temp.groupby(['order_date', 'category_name'], as_index=False)['order_id'].count()
    category_orders_day.rename({'order_id':'Category_orders'},axis=1, inplace=True)


    req = pd.merge(total_orders_day,category_orders_day, on='order_date')

    for i, row in req.iterrows():
        req.loc[i, 'penetration'] = (row['Category_orders']/row['Total_orders'])*100
    
    df = req.groupby(['category_name'], as_index=False)['penetration'].mean()
    avg_day_to_day_penetration.append(df)
```

```python
pd.DataFrame(np.concatenate(avg_day_to_day_penetration), columns=['Categories','Penetration'])
```
